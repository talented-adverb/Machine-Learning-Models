```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from hmmlearn import hmm
```

```python
gen_model = hmm.CategoricalHMM(n_components=2, random_state=99) #random_state=?
```

```python
gen_model.startprob_=np.array([0.4,0.6])
```

```python
gen_model.transmat_=np.array([[0.7,0.3],[0.5,0.5]])
```

```python
gen_model.emissionprob_=np.array([[0.8,0.1,0.1],[0.5,0.2,0.3]])
```

```python
color,gen_states= gen_model.sample(30000)
```

```python
fig, ax = plt.subplots()
ax.plot(gen_states[:50])
ax.set_title('States over time')
ax.set_xlabel('Color')
ax.set_ylabel('Mood')
fig.show()
```

```python
# plot rolls for the fair and loaded states
fig, ax = plt.subplots()
ax.hist(color[gen_states == 0], label='Happy', alpha=0.5,
        bins=np.arange(7) - 0.5, density=True)
ax.hist(color[gen_states == 1], label='Sad', alpha=0.5,
        bins=np.arange(7) - 0.5, density=True)
ax.set_title('Mood probab wrt color')
ax.set_xlabel('No. of days')
ax.set_ylabel('Mood')
ax.legend()
fig.show()
```
